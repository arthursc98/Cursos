# Regression Plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_can=pd.read_excel('Canada.xlsx',sheet_name='Canada by Citizenship',skiprows=20,skipfooter=2)
df_can.columns.astype(str,copy=False)
df_can.head()
```

```python
df_regre=df_can.copy()
df_regre.drop(['Type','Coverage','OdName','AREA','AreaName','REG','RegName','DEV','DevName'],axis=1,inplace=True)
df_regre=pd.DataFrame(df_regre.sum())
df_regre.reset_index(inplace=True)
df_regre.rename(columns={'index':'year',0:'total'},inplace=True)
df_regre.astype(int,copy=False)
df_regre.head()
```

```python
ax=sns.regplot(x='year',y='total',data=df_regre)
```
