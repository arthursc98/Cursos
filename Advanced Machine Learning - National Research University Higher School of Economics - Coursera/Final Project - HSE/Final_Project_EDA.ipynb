Let's make some tests here

So first i'm about to check if our data have some trend or seasonality but for this we need to know if our series if stationary.<br>
##### Null Hypothesis<br>
$$H_0 = Non-Stationary Series$$<hr>
##### Alternative Hypothesis<br>
$$H_1 = Stationary Series$$

```python
hypothesis_testing_df = df.groupby('date').agg({'sellout':'sum'})
utils.adfuller_test(hypothesis_testing_df, 'sellout')
```

Since our Test Statistic is lower then our critical value we can reject the null hypothesis.

Now let's decompose our series.

```python
plt.rcParams['figure.figsize'] = [12, 8]
multimodel=seasonal_decompose(hypothesis_testing_df.values,model='additive',freq=30)
multimodel.plot()
plt.show()
```

We have a seasonality and a some trend also.

Let's try to fit a SARIMAX Model (First time trying to implement this)

```python
sellout_series = hypothesis_testing_df
```

```python
# Defined the p, d and q parameters to take any value between 0 and 2
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(product(p, d, q))]
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(sellout_series.values, order=param, seasonal_order=param_seasonal, enforce_stationarity=False,
                          enforce_invertibility=False)

            results = mod.fit()

            print('SARIMAX{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            continue
```

```python
mod = SARIMAX(sellout_series.values, order=(1,1,1), seasonal_order=(1,1,1,12), enforce_stationarity=False,
              enforce_invertibility=False)

results = mod.fit()

print(results.summary().tables[1])
```

```python
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

- The Correlogram show us that our residual have lower correlation with lagged version of itself.
- The QQ-Plot show us that our residual slight follows a linear trend of the samples taken from a Normal Distribution $N(0,1)$.
