# Data Wrangling

```python
import pandas as pd
```

```python
df=pd.read_csv('automobile.csv')
df.head()
```

### Dealing with missing values

A missing value can be represented in Python with: 'NaN','?',0 or just a blank cell.<br/>Here we will replace every '?' to 'NaN' so we can drop every row that contains a NaN in price column

```python
import numpy as np
df.replace('?',np.nan,inplace=True)
df.head()
```

```python
missing_data=df.isnull()
missing_data.head()
```

If we want some detailed information on which column have a missing value, we can use the function below.

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

Now we will replace every NaN value on the normalized-losses column to a mean his respective mean.

```python
mean_nl=df['normalized-losses'].astype('float').mean()
mean_nl
```

```python
df['normalized-losses'].replace(np.nan,mean_nl,inplace=True)
df.head()
```

```python
mean_bore=df['bore'].astype('float').mean()
mean_bore
```

```python
df['bore'].replace(np.nan,mean_bore,inplace=True)
df.head()
```

To see which values are present in a particular column, we can use the ".value_counts()" method:

```python
df['num-of-doors'].value_counts()
```

Here we can see that the four doors cars are the most common type. But what if we want to see only the most common type? Well to check that we use the ".value_counts().idmax()" method:

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

Now that we know how to treat numeric values and object using Pandas, let's treat every column that contains a missing value.

```python
mean_stroke=df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan,mean_stroke,inplace=True)
mean_horsepower = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, mean_horsepower, inplace=True)
mean_peakrpm=df['peak-rpm'].astype('float').mean()
df['peak-rpm'].replace(np.nan, mean_peakrpm, inplace=True)
```

So we already treated all the columns, now lets drop every row in price column that contains a missing value and reindex the index column, since we deleted a row, the rest of the numbers on index column will not be automatically reindex.

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

### Data formatting

If we want to convert a column type to any other type that we want, we use the astype function, heres an example bellow.

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
df.dtypes
```

### Data Normalization

Theres several ways to normalize data, here comes the most common ways:

Simple Feature scaling<br/>
N=New, O=Old, M=Max<br/>
## $x_n=\frac{x_o}{x_m}$

```python
#Used with Pandas
df['length']=df['length']/df['length'].max()
df['length'].head()
```

Min-Max<br/>
N=New, O=Old, U=Max, L=Min<br/>
## $ x_n=\frac{x_o-x_l}{x_u-x_l}$

```python
df['length']=(df['length']-df['length'].min())/(df['length'].max()-df['length'].min())
df['length'].head()
```

Z-Score<br/>
N=New, O=Old<br/>
## $x_n=\frac{x_o-\mu}{\sigma}$

```python
df['length']=(df['length']-df['length'].mean())/df['length'].std()
df['length'].head()
```

### Binning

```python
bins=np.linspace(min(df['price']),max(df['price']),4)
group_names=['Low','Medium','High']
df['price-binned']=pd.cut(df['price'],bins,labels=group_names,include_lowest=True)
df.head(15)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.hist(df['price'],bins=bins)
plt.title('price bins')
plt.ylabel('count.')
plt.xlabel('price')
plt.show()
```

### Turning categorical variables into quantitative variables 

```python
fuel_qnt=pd.get_dummies(df['fuel-type'])
```

```python
fuel_qnt.head()
```

```python
df=pd.concat([df,fuel_qnt],axis=1)
df.drop(['fuel-type'],axis=1,inplace=True)
df.head()
```

```python
df=pd.concat([df,pd.get_dummies(df['aspiration'])],axis=1)
df.drop(['aspiration'],inplace=True,axis=1)
df.head()
```

```python
plt.title('Proportion of fuel-types')
plt.ylabel('Count of fuel-types')
fuel_qnt.sum().plot(kind='bar')
```
