We have some types of RNN like many to many(voice recognition), one to many(image captioning), many to one(sentiment analysis).

<b>Problems with RNN's</b><br>
1. Computationally expensive.
2. Sensitive to changes in their parameters.
3. Vanishing Gradient
4. Exploding Gradient

# LTSM (Long-Short Term Memory)

We have 3 gates in our RNN:
1. Write gate: Writing data in the memory cell.
2. Read gate: Read data from the memory cell and sends that data back to the RNN.
3. Forget gate: Maintain/Delete data from the memory cell.(Data that is useless.

Here we can see how it works with a Unfolded LSTM Network.<br>


We also have Stacked LSTM which get's the output from a LSTM Network and run again in another LSTM Network providing us a more deeper and higher accuracy at our model, here we have a representation of it.<br>
