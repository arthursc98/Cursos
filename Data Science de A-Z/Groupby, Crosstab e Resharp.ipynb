# Groupby, Crosstab e Resharp

```python
import numpy as np
import pandas as pd
from os import getcwd
```

```python
file=getcwd()+'\\'+input()
drinks=pd.read_csv(file)
drinks.head()
```

```python
drinks.dtypes
```

```python
drinks.population=drinks.population.str.replace('.','').astype(int)
drinks.head()
```

```python
#Média de bebidas servidas na Europa
drinks[drinks.continent=='Europe'].beer_servings.mean()
```

## Calculando a média de cerveja por cada continente

```python
beer_continent=drinks.groupby('continent').beer_servings.mean()
beer_continent.sort_values(ascending=False)
```

```python
def consuming(a):
    if a<=75:
        return 'Low'
    if a>=151:
        return 'High'
    else:
        return 'Middle'
```

```python
#Criando uma nova coluna para medir a qualificação de consumo
drinks['consuming_quantity']=drinks.beer_servings.apply(consuming)
drinks.head()
```

### Calculando o máximo de cerveja consumido em cada continente

```python
drinks.groupby('continent').beer_servings.max()
```

```python
#Com indexação
drinks.to_csv('drinks_indice.csv',encoding='utf-8',index=True)
```

## Trocando quem será o indice

```python
drinks_3=drinks.set_index('consuming_quantity')
drinks_3.head()
```

```python
#Aplicando Groupby
drinks_3=drinks_3.groupby(level='consuming_quantity')
drinks_3.head()
```

```python
drinks_3.sum()
```

```python
drinks_4=drinks.set_index(['consuming_quantity','continent'])
drinks_4.head()
```

```python
#Aplicando Groupby
drinks_4=drinks_4.groupby(level=['consuming_quantity','continent'])
```

```python
drinks_4.sum()
```

### Fazendo o mesmo porém mais simples

```python
drinks_4.aggregate(np.sum)
```

```python
drinks_4.aggregate([np.sum,np.mean,np.size])
```

```python
drinks_5=drinks.set_index(['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol','continent','consuming_quantity'])
```

```python
drinks_5=drinks_5.groupby(level=['consuming_quantity','continent'])
drinks_5.count()
```

```python
drinks_6=drinks.set_index(['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol','continent','consuming_quantity'])
```

```python
drinks_6=drinks_6.groupby(['consuming_quantity','continent','beer_servings'])
drinks_6.sum()
```

# Demonstrando um Crosstab

```python
file=getcwd()+'\\'+input()
movies=pd.read_csv(file)
movies.head()
```

```python
movies.dtypes
```

```python
movies.columns=movies.columns.str.replace(' ','_')
movies.actors_list=movies.actors_list.str.replace('[','').str.replace(']','').str.replace("u",'')
movies.head()
```

```python
pd.crosstab(movies.genre,movies.star_rating)
```
