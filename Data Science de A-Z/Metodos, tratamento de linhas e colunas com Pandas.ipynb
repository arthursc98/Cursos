```python
#Podemos atribuir uma lista aos nomes das colunas
movies.columns = ['star rating', 'title', 'content rating', 'genre', 'duration', 'actors list']
movies.columns
```

```python
#Podemos renomear as colunas no momento de leitura do arquivo
#Header = Cabeçalho do arquivo
columns_movies=['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']
file='C:\\Users\\Arthur\\Desktop\\Data Science\\movies.csv'
movies=pd.read_csv(file,header=0,names=columns_movies)
movies.head()
```

## Os metodos para renomear até agora foram para renomear uma a uma, porém no trabalho a eficiência é fundamental, então mostraremos agora como renomear todas as colunas(tratar para uso de dot notation)

```python
#Em arquivos com csv é necessário usar str.replace para funções que envolvam o uso de strings
file='C:\\Users\\Arthur\\Desktop\\Data Science\\movies.csv'
movies=pd.read_csv(file)
movies.columns=movies.columns.str.replace(' ','_')
movies.head()
```

## Removendo colunas do Dataframe

```python
#Por boa prática e segurança do dataframe é super normal criarmos variváveis para checar se as mudanças que fazemos nele
#estarão de acordo com o que queremos
#axis=1 Colunas | axis=0 Linhas
movies01=movies.drop('content_rating',axis=1)
movies01.head()
```

```python
#Removendo mais de uma coluna
movies02=movies.drop(['duration','actors_list'],axis=1)
movies02.head()
```

## Removendo linhas do Dataframe

```python
#Caso queira remover mais de uma linha, só passar uma lista
movies03=movies02.drop(0,axis=0)
movies03.head()
```
